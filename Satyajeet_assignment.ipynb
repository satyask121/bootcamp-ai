```python
import pandas as pd
from statistics import mean, mode

# Load the dataset (modify the file path as needed)
file_path = "IPL_dataset.csv"  # Change this to the actual path
df = pd.read_csv("E:\wp\ipl_matches_smalldata.csv")

# Assuming the dataset has a column "Runs" and "Player"
# Sort players by runs and select the top 10
top_players = df.sort_values(by="Runs", ascending=False).head(10)

# Extract the runs column
runs = top_players["Runs"].tolist()

# Compute mean and mode
mean_runs = mean(runs)
try:
    mode_runs = mode(runs)
except:
    mode_runs = "No unique mode"

print(f"Mean Runs: {mean_runs}")
print(f"Mode Runs: {mode_runs}")
```

```python
import pandas as pd
df = pd.read_csv(r"E:\wp\ipl_matches_smalldata.csv")
print(df.columns)
top_10_players = df.sort_values(by="total_runs", ascending=False).head(10)
mean_runs = top_10_players["total_runs"].mean()
mode_runs = top_10_players["total_runs"].mode()[0]  
print("\nMean Runs Scored by Top 10 Players:", mean_runs)
print("Mode Runs Scored by Top 10 Players:", mode_runs)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"E:\wp\ipl_matches_smalldata.csv")
print("Columns in dataset:", df.columns)
player_runs = df.groupby("batter")["batsman_runs"].sum().reset_index()
top_10_players = player_runs.sort_values(by="batsman_runs", ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_players["batter"], top_10_players["batsman_runs"], color='blue')
plt.xlabel("Players", fontsize=12)
plt.ylabel("Total Runs", fontsize=12)
plt.title("Top 10 Players by Runs in IPL", fontsize=14)
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
text = """Cricket is a popular bat-and-ball sport that originated in England and is widely played across the world. The game is typically played between two teams of eleven players each. It involves batting, bowling, and fielding in a dynamic contest. The objective is for the batting team to score as many runs as possible while the bowling team tries to dismiss the batsmen and limit the runs. A match is played in either a limited-overs format or traditional longer formats like Test cricket. The game is known for its unique terminology, with terms such as "overs," "wickets," and "boundary" being central to its play. Some of the most prestigious tournaments include the ICC Cricket World Cup and Indian Premier League (IPL). Cricket has evolved over centuries, and today, it is a global sport with passionate fan bases, especially in countries like India, Australia, England, and Pakistan. It brings people together and promotes sportsmanship and team spirit.
"""
print(text)
words = word_tokenize(text.lower()) 
print(words)
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalpha() and word not in stop_words]
print(filtered_words)
word_freq = Counter(filtered_words)
print(word_freq)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
